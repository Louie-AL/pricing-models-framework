```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/royalty_exchange_deals_revenues.csv')
df.head()
```

```python
# make a 2D scatter plot of year of the deal and the multiplier
plt.scatter(df['age'], df['multiplier'])
plt.xlabel('Age')
plt.ylabel('Multiplier')
plt.title('Age vs Multiplier')
plt.show()
```

```python
plt.scatter(df['year'], df['multiplier'])
plt.xlabel('Year')
plt.ylabel('Multiplier')
plt.title('Year vs Multiplier')
plt.show()
```

```python
# plot the average of multiplier for each year
df.groupby('year')['multiplier'].mean().to_frame().reset_index().plot.scatter(x='year', y='multiplier')
plt.xlabel('Year')
plt.ylabel('Average Multiplier')
plt.title('Average Multiplier vs Year')
plt.show()
```

```python
# box plot of multiplier for each year
df.boxplot(column='multiplier', by='year')
plt.ylabel('Multiplier')
plt.title('Boxplot of Multiplier for each Year')
plt.show()
```
