```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
def f(y, r, a, b):
    return 1 / (a * np.exp(-b * (y - 2)) + 1 - a) * (a * np.exp(-b * (y - 1)) / (1 + r - np.exp(-b)) + (1 - a) / r)
```

```python
df = pd.read_csv('./data/revenues_and_deals.csv')

deal_columns = df.columns[3:]
df_deals = df[deal_columns].copy()
df_deals.drop_duplicates(inplace=True)
df_life = df_deals[df_deals['type'] == 'Life of Rights'].copy()
# Drop deals with age less than 2 years

df_life = df_life[df_life['age'] >= 2]
df_life.reset_index(drop=True, inplace=True)

df_life.head()

df_life = df_life[df_life['multiplier'] < 14]
```

```python
# r = [.08, .085, .09, .095, .1, .105, .11, .115, .12, .125, .13, .135, .14, .145, .15, .155, .16, .165, .17, .175]
r = [.08, .085, .09, .095, .1, .105, .11, .115, .12, .13, .14, .15, .16, .17, .18]

df25 = df_life[(df_life['age'] >= 2) & (df_life['age'] < 5)]
df58 = df_life[(df_life['age'] >= 5) & (df_life['age'] < 8)]  
df8plus = df_life[df_life['age'] >= 8]
```

```python
a = .09
b = .27 * 4
c = 1 - a
suffix = '70'
```

```python
# 2-5 years 
for rate in r:
    for i, row in df25.iterrows():
        df25.loc[i, f'model_multiplier_{rate}'] = f(row['age'], rate, a, b)
    # Calculate the mean squared error
    df25[f'se_{rate}'] = (df25[f'multiplier'] - df25[f'model_multiplier_{rate}']) ** 2
```

```python
# 2-5 years mse
for rate in r:
    print(f'Rate: {rate} - MSE: {df25[f"se_{rate}"].mean()}')
```
